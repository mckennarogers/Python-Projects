```python
import requests
import json
import time

# countries I will be using for data
countries = ["Russia", "US", "Switzerland"]

# creates the loop to get all the information for each country
for country in countries:


    url1 = "https://covid-api.mmediagroup.fr/v1/history?country=" #this is the beginning of the url needed
    url2 = "&status=confirmed" #this is the end of the url needed
    url = url1 + country + url2 #creates the completed url
    
    url_requests = requests.get(url) #gets the url
    
    
    country_data = url_requests.text #reads the url
    
    
    country_dct= json.loads(country_data) #loads the url
    
    print("Country: ", country) 



#determines the keys from the data
    key1 = "All"
    key3 = "country"
    key2 = "dates"
    
    
#creates lists (and dictionary) that will be used later to store data    
    total_cases = []
    new_cases = []
    new_dates = []
    monthly_ct = {}
    
    
     #appends the dates into the total_cases list   
    for date in country_dct[key1][key2]:
        new_dates.append(date)
        total_cases.append(country_dct[key1][key2][date])
        
      
    #appends the data (after the math has been done) to the new_cases list
    for i in range(len(total_cases)-1):
        case_difference = total_cases[i] - total_cases[i + 1] 
        new_cases.append(case_difference)
        
    
       
    #finds the average of new cases
    average_new_cases = sum(new_cases)/ len(new_cases) 
    
    print("This is the average of new cases: ", average_new_cases) 
    
    #findes the highest new case
    highest_new_case = max(new_cases)
    
    #finds the date of the highest new cases
    new_cases_index = new_cases.index(highest_new_case)
    
    print("This is the highest day of new cases: ", new_dates[new_cases_index]) #need to print for assignment output
    
    #loops through the dates to find what day is the last with no new cases
    for i in range(len(new_cases)-1):
        no_new_confirmed = new_cases[i]
        if no_new_confirmed == 0: 
            date_no_new_confirmed = new_dates[i]
            print("This is the date with no new confirmed: ", date_no_new_confirmed)
            break
    
    #find the count of new cases for each month
    for i in range(len(new_cases)):
        if new_dates[i][:7] in monthly_ct.keys(): #to find just the characters of year and month
            monthly_ct[new_dates[i][:7]]+=new_cases[i]
        else:
            monthly_ct[new_dates[i][:7]]=new_cases[i]
            
    #creates new lists
    month_tot = []
    months = []
    
    #find the month for each date and append it to the corresponding list
    for month in monthly_ct.keys(): 
        month_tot.append(monthly_ct[month])
        months.append(month)
    
    #find the max month
    highest_month_index = month_tot.index(max(month_tot))
    
    #find the min month
    lowest_month_index = month_tot.index(min(month_tot))
    
    
    
    date_high_new = months[highest_month_index]
    date_low_new = months[lowest_month_index]
    
    
    print("Month with the highest new case count: ", date_high_new)
    print("Month with the lowest new case count: ", date_low_new)
    
    #create the dictionary of the newly found data to be put into a json
    final_dct = {}
    
    #putting all this information into the dictionary above
    final_dct["Country Name: "] = country #repeat this for every output required
    final_dct["Average number of new cases: "] = average_new_cases
    final_dct["Most recent date with no new cases: "] = date_no_new_confirmed
    final_dct["Month with the highest new case count: "] = date_high_new
    final_dct["Month with the lowest new case count: "] = date_low_new
    
    #creating the json with the dictionary and information above
    file = open("/home/ubuntu/environment/hw3/" + country + ".json", "w")
    file.write(str(final_dct))
```
