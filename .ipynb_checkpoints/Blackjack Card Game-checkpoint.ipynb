```python
import random #imported to calculate dealer score

from DeckOfCards import DeckOfCards #imported to get the deck created in DeckOfCards


play = "y" #creates a variable to play the game


while play == 'y': #creates the loop to play the game

    my_deck = DeckOfCards() #variable to use the deck 

    print("Let's Play Black Jack!")
    
    print("Deck before Shuffle: ")
    
    my_deck.print_deck() #prints the deck of cards
    
    print("------")
    
    my_deck.shuffle_deck() #shuffles the deck of cards
    
    print("Shuffled Deck: ")
    
    my_deck.print_deck() #prints the shuffled deck
    
    print("-----")
    
    
    card1 = my_deck.get_card() #creates variable for first card
    card2 = my_deck.get_card() #creates variable for second card
    
   
    current_points=card1.val+card2.val #calculates card score
    
   
    print(card1) #prints first card
    print(card2) #prints second card
    

    print("Your current score is: ", current_points) #prints players score
    
    print("------")
    
    
    hit = input("Would you like to hit? y or n.  ") #creates the user input
    hit_lower = hit.lower() #makes user input lower case
    
    keep_going = 1 #variable to help determine dealer output
    
    while hit_lower != 'y' and hit_lower != 'n': #while loop to ensure user input is y or n
        user_input = input("Please enter y or n ")
        hit_lower = user_input.lower()
    
    print("-----")
    
    while hit_lower == "y": #loop to continue playing game (user inputs y)
        
        newcard = my_deck.get_card() #outputs the new card
        
        print("Your next card is: ", newcard)
        
        current_points += newcard.val #calculates new score
        
        if current_points == 21: #this if elif statement figures out if the player won or not
            print("You got Black Jack! You win!! ")
            keep_going = 0 #for dealer logic
            break #leaves game because player won
        
        elif current_points > 21: 
            print("You busted! You lose.")
            print("Your total is: ", current_points) 
            keep_going = 0 #for dealer logic
            break #leaves loop because player lost
            
        print("Your total is: ", current_points) #prints ending score
        print("------")
        
        hit = input("Would you like to hit? y or n ")
        break #breaks out of the loop
        
        
       
         
    dealer = random.randint(17,23) #creates random score for dealer
    
    print("Dealers Score: ", dealer)
    
    
        
    
    while keep_going == 1: #this loop creates the dealer logic
        if dealer > 21:
            print("Dealer busted. You win!")
        elif dealer == 21:
            print("The dealer got Blackjack! You lose! ")
        else:
            
            if dealer < current_points and current_points > 21:
                print("You Lose")
            elif dealer < current_points: 
                print("You beat the dealer!")
            elif dealer == current_points: 
                print("Tie goes to the house. You lose!")
            else:
                print("You Lose")
        break #leaves the loop
    
    if hit_lower == "n":
        print("-----")
        
    
    play = input("Would you like to play again? y or n.  ")  #creates user input to play again
        
print("Thanks for playing! Have a great day!")
```
